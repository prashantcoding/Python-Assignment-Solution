```python
# Que- Write python program to perform arithmetic operations on two numbers:
def arthmetic_operation(a,b,operation):
  if(operation=='+'):
    return a+b;
  elif(operation=='-'):
    return a-b;
  elif(operation=='/'):
    return a/b;
  else:
    return a*b;
if __name__=="__main__":
  a=int(input());
  b=int(input());
  op=input();
  print(arthmetic_operation(a,b,op));
```

```python
#3.Implement Following Sphere Formulas:
#a.Diameter of Sphere 
#b.	Circumference of Sphere
#c.	Surface area of sphere
#d.	Volume of sphere

def sphere_formula(r):
  print("Radius of the Sphere is")
  Diameter=2*r;
  print(Diameter)
  print("Cirumference area of sphere")
  Circum=2*3.14*r;
  print(Circum)
  print("Surface area of Sphere")
  Surface_Area=4*3.14*r*r;
  print(Surface_Area)
  print("Volume of Sphere")
  Volume=4/3*3.14*(r**3);
  print(Volume);
if __name__=='__main__':
  radius=int(input());
  sphere_formula(radius); 
```

```python
#4.Implement Following Cylinder Formulas:

#a.	Circumference of cylinder
#b.	Surface area of cylinder
#c. Total surface area of Cylinder
#d.	Volume of cylinder

def Cylinder_formula(r,h):
  print("Cirumference  of cylinder")
  Circum=2*3.14*r;
  print(Circum)
  print("Curved surface area of Cylinder")
  Curved_surface=2*3.14*r*h;
  print( Curved_surface)
  print("Total surface area of Cylinder")
  Surface_Area=2*3.14*r*(h+r);
  print(Surface_Area)
  print("Volume of Cylinder")
  Volume=3.14*r*r*h;
  print(Volume);
if __name__=='__main__':
  radius=int(input());
  height=int(input());
  Cylinder_formula(radius,height); 
```

```python
#5.Implement Following Cone Formulas:
#a.	Slant height of cone
#b.	Curved surface area of cone
#c.	Total surface area of cone
#d.	Volume of cone


def Cone_formula(r,h):
  print("Slant height of cone ")
  Slant_Height=(r**r+h**h)**0.5;
  print(Slant_Height)
  print("Curved surface area of cone")
  Diagonal=3.14**r**Slant_Height;
  print(Diagonal)
  print("Total Surface area of cone")
  Surface_Area=3.14*r*(r+Slant_Height);
  print(Surface_Area)
  print("Volume of cone")
  Volume=3.14*r*r*(h/3);
  print(Volume);
if __name__=='__main__':
  r=int(input());
  h=int(input()); 
  Cone_formula(r,h); 
```

```python
#6.Implement Following Sphere Formulas:
#a.Perimeter to Cuboid 
#b.Length of longest diagonal of a cuboid
#c.Total suface area of cuboid
#d.Volume of Cuboid 

def Cuboid_formula(l,b,h):
  print("Perimeter to Cuboid ")
  Peri=4*(l+b+h);
  print(Peri)
  print("Length of longest diagonal of a cuboid")
  Diagonal=((l**l)+(b**b)+(h**h))**0.5;
  print(Diagonal)
  print("Surface area of Sphere")
  Surface_Area=2*(l*b+b*h+h*l);
  print(Surface_Area)
  print("Volume of Sphere")
  Volume=l*b*h;
  print(Volume);
if __name__=='__main__':
  l=int(input());
  b=int(input());
  h=int(input()); 
  Cuboid_formula(l,b,h); 
```

```python
#Q7. Implement compound interest.
def ci(p,r,n,t):
  return p*(1+r/n)**(n*t);
if __name__=='__main__':
  p=int(input());
  r=int(input());
  n=int(input()); 
  t=int(input()); 
  print(ci(p,r,n,t));  
```
