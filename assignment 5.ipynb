```python
#Q5. Write a Python program to count the number of even and odd numbers from a series of numbers(tuple).
def odd_even(tup):
  even_count=0;
  odd_count=0;
  for i in tup:
    if(i%2==0):
      even_count=even_count+1;
    else:
      odd_count=odd_count+1;
  print("even count")
  print(even_count)
  print("odd count")
  print(odd_count)

tuple1 = (0, 1, 2, 3);
odd_even(tuple1)
```

```python
#Q8.Write a Python Program to get the fibonacci Series between 0 to 50

def recur_fibo(n):
   if n <= 1:
       return n
   else:
       return(recur_fibo(n-1) + recur_fibo(n-2))
nterms = 10
if nterms <= 0:
   print("Plese enter a positive integer")
else:
   print("Fibonacci sequence:")
   for i in range(nterms):
       print(recur_fibo(i))
```

```python
#Q9. Write a Python program which iterates the integers from 1 to 50. 
#For multiples of three print "Fizz" instead of the number and for the multiples of five print "Buzz". For numbers which are multiples of both three and five print "FizzBuzz".
for i in range(0,51):
  if (i%3==0 and i%5!=0):
    print("Fizz")
  elif (i%5==0 and i%3!=0):
    print("Buzz")
  elif(i%3==0 and i%5==0):
    print("FizzBuzz")
  else:
    print(i)

```
